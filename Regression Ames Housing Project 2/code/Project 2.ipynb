## 1. Methodology

Data cleaning is necessary in preparing the data for analysis and subsequent modeling. The approach taken on our training and test set is as follows:

Converting all the column names to **standardised lower case with no spacing**  

**Imputed missing and null values**: It is not advisable to drop rows or columns with missing data to prevent data leakage if the dataset is not huge. Hence, it would make better sense to fill in missing and null values by using logical deductions based on the intepretations from the data dictionary. For example, when imputing  missing values in garage area by replacing with the mean or mode based on the appropriateness of the attribute

**Box-plots**: Reviewing box-plots for all categorical, nominal data and dropping attributes with dominating class or ones with similar median and little variance across classes.

**Correlation heatmap, Scatter plot, Multicollinearity Analysis**: Heatmap for numerical variables is constructed with low contributing variables dropped. Scatterplot is used to review to understand the distribution of variables and outliers are addressed.
Multicollinearity analysis is also performed and for variables with high collinearity, the one with higher correlation is retained and the other dropped.

**Feature engineering**: Variables are relooked at and some of the classes within the same variables are collapsed for better data management. For instance, for MS zoning, we have recategorised 8 classes into 2 by mapping the data into residential (R) and non-residential (NR)

**One-hot encoding for categorical nominal variables**: These variables were converted into series of binary variables that represent whether a particular feature was present in a given property or not. 

**Train-Test Split and Scaling**: The train set was train-test split for validation and scaled before processing by regression models.

**Regression Models**: Train set was processed by 3 models - Linear Regression, Ridge CV and LassoCV; and the model with best R2 and RMSE results is selected as final model for testing

**Training of selected model & prediction**:Training model is trained and using test(holdout set) for prediction of housing price

### Lasso Regression

```python
# Instantiation
lasso = LassoCV(alphas=np.logspace(0,5,200),max_iter=50_000, cv=5, random_state=42)

# Data Fitting
lasso.fit(X_train_sc, y_train)
yhat = lasso.predict(X_test_sc)

# Metrics Evaluation (Training data from training set)
print("Training Score: ", cross_val_score(lasso, X_train_sc, y_train).mean())
print("Training RMSE: ", mean_squared_error(y_train, lasso.predict(X_train_sc))**0.5)

#Validation score and validation rsme based on testing data from training set
print("Testing Score: ", r2_score(y_test, yhat))
print("Testing RMSE: ", mean_squared_error(y_test, yhat)**0.5)
```
